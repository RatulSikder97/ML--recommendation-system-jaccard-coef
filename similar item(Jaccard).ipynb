```python
# jaccard function

def jaccardIndex(set1, set2):
    common = len(set(set1).intersection(set(set2)))
    union = len(set(set1).union(set(set2)))
    
    return common/union
```

```python
                
def findSimilarProduct(productId):
    users = []
    for i, userInteraction in rokUserInteaction.iterrows():
        if userInteraction.ITEM_ID == productId:
            users.append(i)
            
    if len(users):    
        items = []
        items = rokUserInteaction.loc[users][rokUserInteaction.loc[users].ITEM_ID != productId].ITEM_ID.values
    
        for item in items:
            result = jaccardIndex(rokItemsData.loc[productId],rokItemsData.loc[item])
            print("Product Id: {product} Rec Id : {item} JAC Index : {jac}".format(product = productId, item = item, jac = result))



        
    
    
        
```
